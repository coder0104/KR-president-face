```python
import matplotlib.pyplot as plt
a = {'family':'serif', 'color': 'g', 'weight' : 'heavy', 'size' : 10}
b = {'family' : 'fantasy', 'color' : 'r', 'weight': 'ultralight', 'size' : 'smaller'}

plt.plot([1,2,3,4,5], [2,4,6,8,10])
plt.xlabel('Xlabel', labelpad=15, fontdict=a)
plt.ylabel("Ylabel", labelpad=30, fontdict=b)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np. arange(5)
years = ['2021', '2022', '2023', '2024', '2025']
values = [100,200,300,400,500]
colors = ['r', 'g', 'y', 'm', 'k']

plt.bar(x, values, color=colors)
plt.xticks(x, years)
plt.xlabel('years')
plt.ylabel('values')

plt.show()
```
